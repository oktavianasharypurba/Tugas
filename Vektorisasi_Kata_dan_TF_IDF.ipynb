```python
#import library yang dibutuhkan
# Import Library yang dibutuhkan

import numpy as np
import pandas as pd
import nltk
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import SnowballStemmer
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/GrayRobert/big-data-project/master/src/main/resources/temp/data/songdata.csv')
```

```python
data
```

```python
texts = data['text'].to_list()
```

```python
texts
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
for i in range(len(texts)):
  texts[i] = texts[i].lower()
```

```python
texts
```

```python
#Menghilangkan character yg tidak jelas
for i in range(len(texts)):
  texts[i] = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,*â€“']", "", texts[i])
```

```python
#Menghilangkan Stopwords dan Juga Stemming
en_stop_words = stopwords.words('english')
stemmer = SnowballStemmer('english')

for i in range(0, len(texts)):
  tokens = word_tokenize(texts[i])
  filtered_sentence = [w for w in tokens if not w in en_stop_words]
  text_clean = [stemmer.stem(word) for word in filtered_sentence]
  texts[i] = ' '.join(text_clean)
```

```python
#Membuat TF.IDF Vectorizer
vectorizer = TfidfVectorizer()
```

```python
#Mengubah text yg sudah dipreprocessing kedalam bentuk TF.IDF
X = vectorizer.fit_transform(texts)
```

```python
#Mentransform Query Menjadi bentuk Vector TF.IDF
query = "begin tell story great love sweet"
query_vec = vectorizer.transform([query])
```

```python
#Menghitung Cosine Similarity Query terhadap Semua Dokumen Yang ada
results = cosine_similarity(X,query_vec).reshape((-1,))
```

```python
#Menampilkan 10 Dokumen dengan Hasil Cosine Similarity Terbesar
for i in results.argsort()[-10:][::-1]:
    print( '\nScore Cosine Similarity : ', results[i], '\n', data.iloc[i,0],"--",data.iloc[i,1])
```
